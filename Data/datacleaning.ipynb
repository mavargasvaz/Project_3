```python
import pandas as pd
import json

# Load the CSV file
csv_filename = 'overdose_DB.csv'
df_csv = pd.read_csv(csv_filename)

# Load the JSON file
json_filename = 'states3.json'
with open(json_filename, 'r') as json_file:
    json_data = json.load(json_file)

# Convert JSON data to DataFrame
df_json = pd.DataFrame(json_data)
df_json = df_json.rename(columns={'state': 'Jurisdiction'})

# Perform the join operation on the 'Jurisdiction' column
df_merged = pd.merge(df_csv, df_json, on='Jurisdiction', how='left')

# Save the merged DataFrame to a new CSV file
output_csv_filename = 'merged_data_real.csv'
df_merged.to_csv(output_csv_filename, index=False)

# Display the merged DataFrame
df_merged

```
